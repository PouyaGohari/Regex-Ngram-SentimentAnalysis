##### Normalizing

With re library functions

```python
import unicodedata
def normalizing_data(content):
    content = content.lower()
    content = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', content)
    content = re.sub(r'www(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', content)
    content = re.sub(r'[^\w\s]', '', content)
    content = re.sub(r'<[^>]+>', '', content)
    content = re.sub(r'\s+', ' ', content).strip()
    content = unicodedata.normalize('NFKD', content).encode('ascii', 'ignore').decode('utf-8')
    return content
with open('normalized_with_my_method_All_Around_the_Moon.txt', 'w') as file:
    file.write(normalizing_data(content))
```

##### Create a class for my model ngram

```python
class ngram:
    def __init__(self, n, data_set):
        self.data = data_set
        self.n = n
    
    def tokenize_data(self):
        self.tokenized = word_tokenize(self.data)
    
    def frequency(self):
        subsequences = []
        for i in range(len(self.tokenized)- self.n + 1):
            subsequence = tuple(self.tokenized[i:i+self.n])
            subsequences.append(subsequence)
        return Counter(subsequences)
    
    def vocab_size(self):
        return len(set(self.tokenized))

    def add_alpba_probability(self, freq, n, alpha = 0.001):
        vocab_size = len(set(self.tokenized))
        probs = {}
        self.n = n-1
        prev_gram_freq = self.frequency()
        self.n = n
        for key,value in freq.items():
            probs[key] = ((value + alpha)/(prev_gram_freq[key[:-1]] + (alpha*vocab_size)))
        return probs
    def complete_sentence_backoff(self, inc_sentence, method, min_lenght=10):
        inc_sentence = normalizing_data(inc_sentence)
        inc_sentence = inc_sentence.translate(str.maketrans('', '', string.punctuation))
        tokens = word_tokenize(inc_sentence)
        current_token = []
        sentence = tokens[:]
        i = self.n
        real_n = self.n
        while len(sentence) < min_lenght + len(tokens):
            candidates = []
            while not candidates:
                current_token = sentence[-i+1:]
                self.n = i
                freq = self.frequency()
                probs = self.add_alpba_probability(freq, i)
                candidates = [(pair, prob) for pair,prob in probs.items() if list(pair[:i-1]) == current_token]
                i = i - 1
            self.n = real_n
            i = self.n
            if method == 'max':
                chosen_token = max(candidates, key= lambda x: x[1])[0][-1]
            else:
                weights = [0 for _ in range(len(candidates))]
                for k in range(len(candidates)):
                    weights[k] = candidates[k][1]
                chosen_token = random.choices(candidates, weights=weights, k=1)[0][0][-1]
            sentence.append(chosen_token)
        return ' '.join(sentence)
    
unigram_model = ngram(1, tarzan_content)
unigram_model.tokenize_data()
print(unigram_model.vocab_size())
```

```python
from nltk import ngrams, word_tokenize, FreqDist
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
path = '/media/pouya/New Volume1/Master SE/Term 2/NLP/CA\'s/CA1/data/google_play_store_apps_reviews.csv'
review_data = pd.read_csv(path)
review_data.head()
```

```python
train_data, test_data = train_test_split(review_data, test_size=0.2, random_state=42)
```

```python
def get_ngrams(text, n):
    tokens = word_tokenize(text)
    return list(ngrams(tokens, n))

def train_ngram(data, n):
    positive_ngrams = []
    negative_ngrams = []
    for index, row in data.iterrows():
        grams = get_ngrams(row['review'], n)
        if row['polarity'] == 1:
            positive_ngrams.extend(grams)
        else:
            negative_ngrams.extend(grams)
    positive_freq = FreqDist(positive_ngrams)
    negative_freq = FreqDist(negative_ngrams)
    return positive_freq, negative_freq
        
```

```python
n = 2
positive_freq, negative_freq = train_ngram(train_data, n)
```

```python
def predict_polarity(review, positive_freq, negative_freq, n):
    grams = get_ngrams(review, n)
    positive_prob = sum(np.log(positive_freq.freq(gram) + 1e-6) for gram in grams)
    negative_prob = sum(np.log(negative_freq.freq(gram) + 1e-6) for gram in grams)
    label = 1 if positive_prob >= negative_prob else 0
    return label

def test_ngram(test_data, positive_freq, negative_freq, n):
    correct_predictions = 0
    predicted_labels = []
    for _,row in test_data.iterrows():
        predicted = predict_polarity(row['review'], positive_freq, negative_freq, n)
        predicted_labels.append(predicted)
        if predicted == row['polarity']:
            correct_predictions += 1
    return correct_predictions/len(test_data), predicted_labels
```

```python
print(test_ngram(test_data, positive_freq, negative_freq, n)[0])
test_ngram(test_data, positive_freq, negative_freq, n)[1]
```

##### with built-in fucntions

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
```

```python
actual = test_data['polarity'].tolist()
predicted = [predict_polarity(row['review'], positive_freq, negative_freq, n) for index,row in test_data.iterrows()]
print("Precision:", precision_score(actual, predicted))
print("Recall:", recall_score(actual, predicted))
print("F1 Score:", f1_score(actual, predicted))
print("Confusion Matrix:\n", confusion_matrix(actual, predicted))
```

```python
n = [i for i in range(1, 8, 1)]
precision_scores = []
for x in n:
    positive_freq, negative_freq = train_ngram(train_data, x)
    predicted = [predict_polarity(row['review'], positive_freq, negative_freq, x) for index,row in test_data.iterrows()]
    precision_scores.append(precision_score(actual, predicted))
print(precision_scores)
```
